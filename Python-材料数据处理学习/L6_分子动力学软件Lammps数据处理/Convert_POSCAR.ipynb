```python
with open("Class-6.data",'r') as reader:
	all_content = reader.readlines()
	#print(all_content)
	for index,line in enumerate(all_content):  #enumerate 迭代可以告诉你迭代的是第几行
		if "atoms" in line:   #判断atoms是否在这一行里
			atomnus=int(line.split()[0])
		if "atom" in line and "types" in line:
			typesnums = int(line.split()[0])
		if "xlo" in line and "xhi" in line:
			x = float(line.split()[1])-float(line.split()[0])   #x方向的长度
		if "ylo" in line and "yhi" in line:
			y = float(line.split()[1])-float(line.split()[0])
		if "zlo" in line and "zhi" in line:
			z = float(line.split()[1])-float(line.split()[0])
		if "xy" in line and "xz" in line and "yz" in line:
			xy = float(line.split()[0])
			xz = float(line.split()[1])
			yz = float(line.split()[2])
		if "Atoms" in line:   #Atoms 在第几行，从Atoms行后开始读下去
			atomsbegin = index
	print(atomnus,typesnums,x,y,z,xy,xz,yz)
newlattice = [[x,0,0],[xy,y,0],[xz,yz,z]]
print(newlattice)

for i in range(atomsbegin+1,len(all_content)):
	if all_content[i].strip("\r\n").strip("\n") != "":  #真正有数据的行是第几行
		atomsbegin = i  #从第17行开始读
		break
atoms = []
for i in all_content[atomsbegin:atomsbegin+atomnus]:
    atoms.append(('c',float(i.split()[2]),float(i.split()[3]),float(i.split()[4]))) 
print(atoms)   

element = [i[0] for i in atoms]  #取所有原子的元素信息，构成一个列表
#print(element)
element_s = set(element)  #用set方法进行去重复
print(element_s)
""" 
如果xsd中的元素有混乱的情况['Cu','O','Cu'] ， 还需要对之前的atoms进行排序，
def sortatom(i)
	return i[0]
atoms.sort(key=sortatom) #对原子信息进行原位排序，排序依据是它的元素信息
"""

def mysorted(i):  #自定义排序规则，返回的是传入的元素在原来列表中的顺序
	return element.index(i)
element_s = sorted(element_s,key=mysorted) #对去重后的元素列表进行排序，
print(element_s)
element_num = []
for i in element_s:
	element_num.append(element.count(i)) #对重排后的元素种类列表迭代，得到每个元素的原子数目
print(element_num)

		
	#for index, line in enumerate(all_content):

```
