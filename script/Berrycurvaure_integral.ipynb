```python
""" 通过wannier-tools 计算得到Berry 曲率，提取数据对其进行积分，方法二，测试文件 """
file=open("test1.dat",'r') 

all_data=file.read()
file.seek(0)
all_data_lines = file.readlines()
all_data_s =all_data.split(" \n")  #按空格和换行符进行分割，把每个段落都进行分割，其中第一乱文字也被分割,形成一个大列表
print(all_data_s)

```

```python
eachline=[]  
eachline1=[]
for i in all_data_s:  #对每一段进行迭代。
    line=i.split('\n') #对每一段\n 进行切割，分成每一行。line 表示一个列表，并且每一行为一个元素。
    for j in line:    #对该段的每一行（元素）进行迭代。
        if 'Please' not in  j and 'kx' not in j  : #筛选含有字母的前行。
            eachline.append(j)  #所选的每一行做为一个元素添加到空列表eachline里
print(eachline)
```

```python
for i in eachline:  #对该列表每一行（元素）进行再次迭代，目的是去掉空格'' and ' '
    if '' not in i.split(',') and ' ' not in i.split(','): #首先要把'' and ' '转化成单独的列表，才能判断用if in not 来判断。通过split方式进行对',' 进行切分。
        eachline1.append(list(map(float,i.split())))   #对每一行（一个大字符）通过空格进行切分后，形成多个字符元素，然后再通过float方式转化成浮点数。
print(eachline1)
```

```python
import numpy as np
new_data=np.array(eachline1)   #把一个二维列表转化成形似相类似的数组。
print(new_data)
```
