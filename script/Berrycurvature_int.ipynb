```python
""" wannier-tools Berry  """

file=open("Berrycurvature.dat",'r')

all_data=file.read()
file.seek(0)
all_data_lines = file.readlines()
#print(all_data_lines)
all_data1 =all_data.split(" \n")
eachline=[]
eachline1=[]
for i in all_data1[1:]:   #
    if i.split("\n") !='':
        lines = i.split("\n")    #
        for j in lines:   #
            line_num=j.split(",")   # '' and ' ' 
            if '' not in line_num and ' 'not in line_num:  # '' and ' ' 
                eachline.append(line_num)  #eachline
for i in eachline[1:]: #
    for j in i:  #()split
        eachline1.append(list(map(float,j.split())))  #

import numpy as np
new_data=np.array(eachline1)     #list numpy
#print(new_data)
value = np.sum(new_data,axis=0)
#print(value)   
#         #for line in lines:
```
