```python
""" 通过wannier-tools 计算得到Berry 曲率，提取数据对其进行积分，方法一 """

file=open("Berrycurvature.dat",'r')

all_data=file.read()
file.seek(0)
all_data_lines = file.readlines()
#print(all_data_lines)
all_data1 =all_data.split(" \n")
eachline=[]
eachline1=[]
for i in all_data1[1:]:   #对每一段进行迭代，排除含有字母的第一行
    if i.split("\n") !='':
        lines = i.split("\n")    #把其中每一段，按照每一行（一个元素）进行分割，一个列表（一段），每行（一个元素）
        for j in lines:   #对每一行（一个元素）进行迭代，
            line_num=j.split(",")   #为了排除空格 '' and ' ' （它们此刻是元素），把他们元素转化成列表。
            if '' not in line_num and ' 'not in line_num:  #判断 '' and ' ' 是否在这列表中，从而进行筛选。
                eachline.append(line_num)  #对所要选的行数据，添加到空列表eachline中，形成一个大列表，包含每一行的列表。
for i in eachline[1:]: #从含有文字的下一行，开始进行每一行（元素）进行迭代。
    for j in i:  #然后对一行一个(列表)进行迭代，虽然只有一个元素，目的是提取文本格式，才能进行split
        eachline1.append(list(map(float,j.split())))  #对每一行数据通过空格进行分割，然后再转化为浮点格式。

import numpy as np
new_data=np.array(eachline1)     #把list 浮点格式转化为numpy，从而进行数据分析。
#print(new_data)
value = np.sum(new_data,axis=0)
#print(value)   
#         #for line in lines:
```
